```python
from scipy.spatial import distance as dist
arr1 = [[3,4],[1,2],[5,6],[10,10]]
arr2 = [[1,2],[3,5],[3,4],[10,10]]

d = dist.cdist(arr1, arr2)
rows = d.min(axis=1).argsort()
cols = d.argmin(axis=1)[rows]

print(d)
print(cols)
print(rows)
```
